#### QUESTION:
Make an array having 10 equal interval numbers which will represent points on x-axis.On the y axis plot 4 different operations graph in single plot with different shapes marked for comparision?
Following 3 are the operations:
(A) Plot x (Straight Line)
(D) Plot x VS (x)^2
(C) Plot x VS (x)^2.5
(D) Plot x VS (x)^3

#### NOTE:
(A) Width of the line must be thicker than default.
(B) Use different color and design for every line.
(C) Graph must contains Legend and Grid.


```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11,1)
print("X-axis Values:",x)
```

```python
plt.plot(x,x,label='Numbers',linewidth=3.0)
plt.plot(x,x**2,'b--',label='Power 2 of the numbers',linewidth=3.0)
plt.plot(x,x**2.5,'g^--',label='Power 2.5 of the numbers',linewidth=3.0)
plt.plot(x,x**3,'rs-',label='Power 3 of the numbers',linewidth=3.0)
plt.grid()
plt.ylabel('squares')
plt.xlabel('numbers')
plt.title('Power Graph')
plt.legend()
plt.show()
```
